```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 1. Criar tabela de conting√™ncia (dia da semana x tipo de acidente)
tabela = pd.crosstab(df_pf_sin["dia_semana"], df_pf_sin["tipo_acidente"]) #cada c√©lula mostra quantos acidentes de um certo tipo aconteceram em um determinado dia da semana.

# 2. Plotar heatmap de frequ√™ncias
plt.figure(figsize=(12,6)) #Cria uma figura do gr√°fico com tamanho 12x6 polegadas.
sns.heatmap(
    tabela,  #dados de entrada (a tabela de conting√™ncia
    annot=True, #escreve os valores dentro de cada c√©lula.
    fmt="d", #mostra os n√∫meros como inteiros (d = decimal integer
    cmap="Blues" #cor (quanto maior o valor, mais escuro o tom)
)
plt.title("Frequ√™ncia de Acidentes por Dia da Semana e Tipo de Acidente", fontsize=14)
plt.xlabel("Tipo de Acidente")
plt.ylabel("Dia da Semana")
plt.show()  #configura o t√≠tulo

# 3. Teste Qui-quadrado
chi2, p, dof, expected = chi2_contingency(tabela)
#chi (valor do qui-quadrado calculado)
#p-valor do teste (indica signific√¢ncia estat√≠stica).
#graus de liberdade.
#expected: valores esperados se as vari√°veis fossem independentes (n√£o exibidos no print, mas retornados).

print("üîπ Resultado do Teste Qui-quadrado")
print(f"Qui-quadrado: {chi2:.2f}")
print(f"p-valor: {p:.4f}")
print(f"Graus de liberdade: {dof}")

# 4. Interpreta√ß√£o autom√°tica
if p < 0.05:
    print("\n‚úÖ Existe evid√™ncia estat√≠stica de que DIA DA SEMANA e TIPO DE ACIDENTE est√£o relacionados.")
else:
    print("\n‚ùå N√£o h√° evid√™ncia estat√≠stica suficiente para afirmar que h√° rela√ß√£o entre DIA DA SEMANA e TIPO DE ACIDENTE.")
```
